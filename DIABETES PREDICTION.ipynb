# DATA PREPROCESSING

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = diabetesDF.drop('class', axis=1).values
X.shape
```

```python
y = diabetesDF['class']
```

```python
sns.countplot(y)
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(y_test.tolist())
```

```python
print(confusion_matrix(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix = confusion_matrix(y_test,y_pred)
ax = sns.heatmap(confusion_matrix, cmap='flare',annot=True, fmt='d')
plt.xlabel("Predicted Class",fontsize=12)
plt.ylabel("True Class",fontsize=12)
plt.title("Confusion Matrix",fontsize=12)
```

# NEURAL NETWORK MODEL 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = diabetesDF.drop('class', axis=1).values
```

```python
y = diabetesDF['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size = 0.3,
                                                        stratify = y,
                                                         random_state = 99)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(8,activation='relu',input_shape=(16,)))
model.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
model.compile(optimizer='adam',
                     loss='sparse_categorical_crossentropy',
                     metrics='accuracy')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, \
                     batch_size = 8, epochs= 50, \
                     verbose=2, \
                     validation_split=0.2)
```

```python
accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']
plt.plot(accuracy, label='Training Set Accuracy')
plt.plot(validation_accuracy, label='Validation Set Accuracy')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy Across Epochs')
plt.legend()
```

```python
loss = history.history['loss']
validation_loss = history.history['val_loss']
plt.plot(loss, label='Training Set Loss')
plt.plot(validation_loss, label='Validation Set Loss')
plt.ylabel('Loss')
plt.title('Training and Validation Accuracy Across Epochs')
plt.legend()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred1 = model.predict(X_test)
y_pred1 = y_pred1.argmax(axis=1)
```

```python
confusion_matrix = confusion_matrix(y_test,y_pred1)
ax = sns.heatmap(confusion_matrix, cmap='flare',annot=True, fmt='d')
plt.xlabel("Predicted Class",fontsize=12)
plt.ylabel("True Class",fontsize=12)
plt.title("Confusion Matrix",fontsize=12)
```

```python
print(classification_report(y_test,y_pred1))
```

# MODEL USING FEATURE SELECTION (FILTER METHOD)

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X = diabetesDF.drop('class', axis=1)
```

```python
X.columns
```

```python
y = diabetesDF['class']
```

```python
chi_best = SelectKBest(score_func=chi2, k=12)
k_best = chi_best.fit(X, y)
```

```python
print(k_best.scores_)
```

```python
scores = pd.DataFrame(k_best.scores_)
cols = pd.DataFrame(X.columns)
```

```python
feature = pd.concat([cols,scores],axis=1)
feature.columns = ['column', 'scores']
```

```python
X1 = feature.nlargest(12, 'scores')
```

```python
X1 = diabetesDF[['Polydipsia','sudden weight loss','partial paresis',
             'Irritability','Polyphagia','Age',
             'visual blurring', 'Polyuria', 'Gender',
                 'Alopecia', 'weakness','Genital thrush']]
y = diabetesDF['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y, 
                                                        test_size = 0.3,
                                                         random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# RANDOM FOREST MODEL 2

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```
